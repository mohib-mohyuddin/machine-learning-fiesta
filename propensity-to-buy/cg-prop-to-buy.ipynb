# 1. Exploratory Data Analysis

```python
#First, let's create a new variable "converted_yes" that returns 1 if converted, 0 otherwise

cg_organic['converted_yes'] = np.where(cg_organic['converted'] == 'yes', 1, 0)

total_converted = cg_organic.converted_yes.sum()
prob_converted = total_converted / len(cg_organic)
print(f'The probability of organically converting to is {prob_converted}')
```

### 2) Generate basic summary statistics for each feature in the data. For numeric variables shown the number of observations, the number of missing values, the number of distinct values, and the mean, min, max, and standard deviation. For non-numeric variables show the number of observations, the number of missing values, the number of distinct values, the most common level, and the least common level

```python
# First, separate the dataframe into numeric and non-numeric data
numeric_data = cg_organic.select_dtypes(include=[np.number])
non_numeric_data = cg_organic.select_dtypes(exclude=[np.number])

# For Numeric Variables
numeric_stats = numeric_data.describe().T  # Basic stats
numeric_stats['missing'] = numeric_data.isna().sum()  # Missing values
numeric_stats['nunique'] = numeric_data.nunique()  # Distinct values

# For Non-Numeric Variables
non_numeric_stats = non_numeric_data.describe(include='all').T  # Basic stats for non-numeric
non_numeric_stats['missing'] = non_numeric_data.isna().sum()  # Missing values
non_numeric_stats['nunique'] = non_numeric_data.nunique()  # Distinct values
non_numeric_stats['most_common'] = non_numeric_data.apply(lambda x: x.value_counts().index[0])
non_numeric_stats['least_common'] = non_numeric_data.apply(lambda x: x.value_counts().index[-1])  # Least common level

# Print or return the statistics
print("Numeric Variables Statistics:")
print(numeric_stats)
print("\nNon-Numeric Variables Statistics:")
print(non_numeric_stats)
```

# 2. Predictive Model

Now that we have done an exploratory analysis, we will build a predictive model using a logistic regression and all the variables in our dataset.

```python
lr = rsm.logistic({"cg_organig_train": cg_organic[cg_organic['training']==1]}, rvar='converted_yes', evar=['GameLevel', 'NumGameDays','NumGameDays4Plus','NumInGameMessagesSent','NumSpaceHeroBadges','NumFriendRequestIgnored','NumFriends','AcquiredSpaceship', 'AcquiredIonWeapon', 'TimesLostSpaceship', 'TimesKilled', 'TimesCaptain', 'TimesNavigator', 'PurchasedCoinPackSmall', 'PurchasedCoinPackLarge', 'NumAdsClicked', 'DaysUser', 'UserConsole', 'UserHasOldOS'])
lr.summary()
lr.plot('vimp')
```

### What are the 5 most important features

As per the permutation plot, NumSpaceHerobades, NumFriendsRequestignored, TimelostSpaceship, GameLevel and NumAdsclicked are the most important feature of this model

```python
lr.plot('pred', incl=['NumSpaceHeroBadges', 'NumFriendRequestIgnored', 'TimesLostSpaceship', 'GameLevel', 'NumAdsClicked'])
```

### Prediction plots of the 5 most important features
As per the predictions plot, we can say:
* Numspaceherobadges : The higher is the number of badges, the more likely the player will convert organically to the campaign.
* Numfriendsrequestignored: The more friends request ignored, the less likely the player will convert to the campaign. However, the decrease in probability is not sharp
* TimesLostSpaceShip: The more spaceship the player has lost, the less likely he will convert to the compaign organically.
* GameLevel: The higher is the game level of the player, the more likely he will convert to the campaign
* NumAdsClicked: The more the player has clicked on previous ads, the more likely he will convert to the campaign

### 1. Train and tune a neural network on the random sample of customers who wereexposed to the ad campaign (i.e., rnd_30k == 1). Select two hyper parameters for your grid to tune on.

```python
NN = rsm.model.mlp(
    data={'ads_campaign':ads[ads['rnd_30k'] == 1]},
    rvar='converted',
    lev='yes',
    evar=['GameLevel', 'NumGameDays','NumGameDays4Plus','NumInGameMessagesSent','NumSpaceHeroBadges','NumFriendRequestIgnored','NumFriends','AcquiredSpaceship', 'AcquiredIonWeapon', 'TimesLostSpaceship', 'TimesKilled', 'TimesCaptain', 'TimesNavigator', 'PurchasedCoinPackSmall', 'PurchasedCoinPackLarge', 'NumAdsClicked', 'DaysUser', 'UserConsole', 'UserHasOldOS'],
    hidden_layer_sizes=(4,),
    mod_type='classification',
)

NN.summary()
```

```python
NN.plot('vimp')
```

```python
from sklearn import metrics
from sklearn.model_selection import GridSearchCV

hls = [
        (1,), (2,), (3,), (4,), (5,),
        (2,1), (2,2), (2,3), (2,4), (2,5),
        (3,1), (3,2), (3,3), (3,4), (3,5),
       ]

learning_rate = ['constant', 'invscaling', 'adaptive']

param_grid = {'hidden_layer_sizes': hls,
              'learning_rate': learning_rate}

scoring = {'AUC': 'roc_auc'}

nn_cv = GridSearchCV(
    NN.fitted, param_grid, scoring=scoring, cv=5, n_jobs=4, refit='AUC', verbose=5
)
```

```python
nn_cv.fit(NN.data_onehot, NN.data.converted)
```

```python
pd.DataFrame(nn_cv.cv_results_).iloc[:, 5:].sort_values("rank_test_AUC")
```

```python
print(nn_cv.best_params_)
print(nn_cv.best_score_.round(3))
```

After training a neural network model using the same features as our previous model, we used GridSearchCV to fine-tune the hyper-parameters.

It seems like the most performing model is with 2 layers, with 3 and 4 nodes respectively. The Learning rate seems to be equivalent across when using these layers. Therefore, we can assume that layers 3 nodes and 4 nodes and a constant learning rate are the hyperparameters to go to.
